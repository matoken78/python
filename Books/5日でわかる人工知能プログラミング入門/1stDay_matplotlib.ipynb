```python
import matplotlib.pyplot as plt

# インライン表示するための宣言
%matplotlib inline#
```

```python
import math
import numpy as np

# x 座標は 0～5π
# y 座標はサイン波
x = np.linspace(0, 5 * math.pi)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python
# title 関数でタイトルを追加
plt.title('Sin Graph')

# 軸名を追加
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

# 凡例を指定
plt.plot(x, y, label='sin')

# 凡例を表示
plt.legend()

plt.show()
```
