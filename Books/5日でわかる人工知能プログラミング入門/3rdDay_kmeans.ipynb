```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], test_size=0.3, random_state=0)
```

```python
import pandas as pd
pd.DataFrame(x_train, columns=iris_dataset.feature_names)
```

```python
pd.DataFrame(x_test, columns=iris_dataset.feature_names)
```

```python
# mglearn は matplotlib のヘルパーモジュール
import mglearn
import matplotlib.pyplot as plt
```

```python
iris_dataframe = pd.DataFrame(x_train, columns=iris_dataset.feature_names)
```

```python
# scatter_matrix は、一連の変数のペアプロットを作成して散布図にしてくれる
from pandas.plotting import scatter_matrix
grr = scatter_matrix(iris_dataframe, c=y_train, figsize=(15, 15), marker='o', hist_kwds={'bins': 20}, s=60, alpha=.8, cmap=mglearn.cm3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(x_train, y_train)
```

```python
# 予測（1つ目のデータ）
import numpy as np
# setosa の がく片の長さ、がく片の幅、花弁の長さ、花弁の幅
x_new = np.array([[5.0, 2.9, 1.0, 0.2]])
prediction1 = knn.predict(x_new)
print(iris_dataset['target_names'][prediction1])
```

```python
# 予測（2つ目のデータ）
x_new = np.array([[6.0, 2.7, 5.1, 1.6]])
prediction1 = knn.predict(x_new)
print(iris_dataset['target_names'][prediction1])
```

```python
# モデルの評価（正解率の算出）
y_pred = knn.predict(x_test)
# 予測結果
print('予測結果')
print(y_pred)
# 正解ラベル
print('正解ラベル')
print(y_test)
# 正解率=予測精度
print('予測精度')
np.mean(y_pred == y_test)
```
