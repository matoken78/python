# ニューラルネットワーク
* 人間の脳にある「ニューロン」という神経細胞を模したプログラムを複数接続した機械学習システムのこと  

# 形式ニューロン
* 人間のニューロンは、外部から様々な電気刺激を受けて興奮し  
やがて別のニューロンに電気信号を出力することで興奮が収束するという性質がある。  
このニューロンを模して作られたのが「形式ニューロン」(人工ニューロン)と呼ばれるプログラム。  
* 形式ニューロンは、「入力(0か1)×重み」の総和が閾値を超えていると1を出力し  
超えていなければ0を出力するという単純なプログラム。  
この計算によって1を出力するか0を出力するかを決める関数を「活性化関数」と呼ぶ。  
「単純パーセプトロン」は、この形式ニューロンを機械学習に利用したもの。

# 単純パーセプトロン
* 形式ニューロンは通常、入力が1か0のどちらかだが、入力を実数に対応させ、かつ  
学習により重みを更新できるように改良したものが「単純パーセプトロン」。
* 単純パーセプトロンは、線形分離可能なデータしか判別できず  
この欠点を「XOR問題」と言う。  
⇒「AND」「OR」「NAND」は判別できる(線形分離可能)が、「XOR」は判別できない(線形分離不可能)。

# 多層パーセプトロン
* OR、AND、NAND の判別をする形式ニューロンを多層に組み合わせると  
XOR の識別ができることがわかっている。  
つまり、形式ニューロン(単純パーセプトロン)を何層も組み合わせることで  
複雑な決定境界を引けるようになる、これが「多層パーセプトロン」。  
* 多層パーセプトロンでは、入力部分を「入力層」、出力部分を「出力層」、  
その間にあるパーセプトロンを「隠れ層」(中間層)と呼び、  
この隠れ層を2重3重と増やすことで複雑な決定境界を引くことができるようになる。  
* このような多層パーセプトロンを利用した機械学習システムを「ニューラルネットワーク」と呼ぶ。

# 多層パーセプトロンの活性化関数
* ニューラルネットワークの活性化関数には、「シグモイド関数」が利用されていたが  
最近は「ReLU(Rectified Linear Unit)関数」が用いられる。  
ReLU関数は、入力が0を超えていれば入力をそのまま出力し、0以下なら0を出力する関数。

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# xの値(-5～5の間で0.1刻み)
x = np.arange(-5.0, 5.0, 0.1)

# yの値(0とxの要素を比較した大きい方)
y = np.maximum(0, x)

# グラフの設定と表示
plt.title("ReLU function")
plt.plot(x, y)
plt.ylim(-0.1, 5.1)
plt.show()
```

ちなみに、入力をそのまま出力する関数を「恒等関数」よ呼ぶ。

```python
x = np.arange(-5.0, 5.0, 0.1)
y = x
plt.title("Identify function")
plt.plot(x, y)
plt.ylim(-5.1, 5.1)
plt.show()
```

ニューラルネットワークで分類を行う際、  
出力層の活性化関数に「ソフトマックス関数」(正規化指数関数)を採用することがある。
ソフトマックス関数は、出力がどのクラスに分類されるのかを、確率で出力するために使う。

```python
x = np.arange(-5.0, 5.0, 0.1)
exp_x = np.exp(x)
sum_exp_x = np.sum(exp_x)
y = exp_x / sum_exp_x

plt.title("Softmax function")
plt.plot(x, y)
plt.show()
```

# 誤差逆伝播(バックプロパゲーション、Backpropagation)
* 単純パーセプトロンでは更新する重みベクトルは1つだが  
多層パーセプトロンの場合、どの順番でパーセプトロンが判断したかによって  
更新する重みが変化する。  
* ニューロンが逆にたどりながら重みを更新するアルゴリズムとしては「誤差逆伝播」が有名。
